## Q3 - Write a code for max pooling and unpooling. You need to store max locations in another matrix.
### Assuming that the stride here is of 2

image.png

```python
import numpy as np
```

```python
arr = np.array([[0.1,0.5,1.2,-0.7],
              [0.8,-0.2,-0.5,0.3],
              [0.4,0.9,-0.1,-0.2],
              [-0.6,0.1,0.5,0.3]])




max_arr = np.zeros((2,2))

print(arr)
```

```python
for i in range(0,2):
  for j in range(0,2):
    temp=[]

    for k in range(i*2,(i+1)*2):
      for l in range(j*2, (j+1)*2):
        temp.append(arr[k][l])


    max_arr[i][j] = max(temp)

```

```python
print(max_arr)
```

```python
loc=[]

for i in range(0,2):
  for j in range(0,2):


    for k in range(i*2,(i+1)*2):
      for l in range(j*2, (j+1)*2):
        if(arr[k][l]==max_arr[i][j]):
          loc.append((k,l))

```

```python
print(loc) #now we have locations of max values
```

### Unpooling

```python
#flattening max pooled array

flat_max_pool=[]

for i in range(0,np.shape(max_arr)[0]):
  for j in range(0,np.shape(max_arr)[1]):
    flat_max_pool.append(max_arr[i][j])


flat_max_pool
```

```python
back_prop_arr = np.zeros((np.shape(arr)[0],np.shape(arr)[1]))

for i in range(0,len(flat_max_pool)):
  back_prop_arr[loc[i][0]][loc[i][1]] = flat_max_pool[i]
```

```python
print(back_prop_arr)
```
