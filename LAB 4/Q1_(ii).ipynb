```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import plot_model
import matplotlib.pyplot as plt
```

```python
model = Sequential()
model.add(Dense(6, input_shape=(6, ), name='dense1'))
model.add(Dense(100, name='dense2'))
model.add(Dense(500, name='dense3'))
model.add(Dense(200, name='dense4'))
model.add(Dense(50, name='dense5'))
model.add(Dense(2, name='dense_output',activation="sigmoid"))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
model.summary()

# # Plot the model
plot_model(model, to_file='model.png')
# # Display the image
data = plt.imread('model.png')
plt.imshow(data)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('heart_disease_dataset_UCI.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X=df[['chol','trestbps','thalach','sex','age','oldpeak']]
y=df[['target','thal']]


#X_train, X_test, y_train, y_test = train_test_split(df[['chol','trestbps','thalach']], df['target'])
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X, y)
```

```python
model.fit(X_train_new, y_train_new,
          epochs=50,
          batch_size=32,
          validation_split=.10,
          verbose=True)
```

```python
# Evaluate the model on the test data
model.evaluate(X_test_new, y_test_new)
```
