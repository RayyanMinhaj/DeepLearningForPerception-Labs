```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import plot_model
import matplotlib.pyplot as plt
model = Sequential()
model.add(Dense(4, input_shape=(3, ), name='dense1'))
model.add(Dense(4, name='dense2'))
model.add(Dense(4, name='dense3'))
model.add(Dense(1, name='dense4',activation="sigmoid"))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
model.summary()

# # Plot the model
plot_model(model, to_file='model.png')
# # Display the image
data = plt.imread('model.png')
plt.imshow(data)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('heart_disease_dataset_UCI.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X=df[['chol','trestbps','thalach']]
y=df[['target']]


#X_train, X_test, y_train, y_test = train_test_split(df[['chol','trestbps','thalach']], df['target'])
X_train, X_test, y_train, y_test = train_test_split(X, y)


```

```python
model.fit(X_train, y_train,
          epochs=50,
          batch_size=32,
          validation_split=.10,
          verbose=True)
```

```python
# Evaluate the model on the test data
model.evaluate(X_test, y_test)
```
